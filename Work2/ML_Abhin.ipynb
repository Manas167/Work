```python
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
```

```python
centers = [[-5], [0]]
X, y = make_blobs(n_samples=1000, centers=centers, random_state=40)
X = sm.add_constant(X)
X = pd.concat([pd.DataFrame(X), pd.DataFrame(np.random.randint(0, 2, len(X)))], 1)
X.columns = ['const', 'one', 0]
print(X.head())

X_expanded = pd.concat([X, X['one']*X[0]], 1)
X_expanded.columns = ['const','one', 0, 'one0']
print(X_expanded.head())
print(y[:5])
```

```python
sm.OLS(y, X_expanded).fit().summary()
```
